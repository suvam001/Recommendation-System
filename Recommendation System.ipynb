```python
import pandas as pd
import numpy  as np
```

```python
credits = pd.read_csv(r"C:\Users\DELL\Downloads\archive\tmdb_5000_credits.csv")
movies = pd.read_csv(r"C:\Users\DELL\Downloads\archive\tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("Credits shape: ", credits.shape)
print("Movies shape: ", movies.shape)
```

```python
renamed_credits = credits.rename(index = str, columns= {"movie_id":"id"})
merged_columns = movies.merge(renamed_credits, on = "id")
merged_columns.head()
```

```python
movies_cleaned = merged_columns.drop(columns = ['homepage', 'status', 'title_x', 'title_y', 'production_countries'])
```

```python
movies_cleaned.head()
```

```python
v=movies_cleaned['vote_count']
R=movies_cleaned['vote_average']
C=movies_cleaned['vote_average'].mean()
m=movies_cleaned['vote_count'].quantile(0.70)
```

```python
movies_cleaned['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
movies_cleaned.head()
```

```python
movies_cleaned[['original_title', 'popularity', 'weighted_average', 'vote_average', 'revenue']].head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled=scaling.fit_transform(movies_cleaned[['popularity', 'weighted_average', 'vote_average', 'revenue']])
movie_normalized_df=pd.DataFrame(movie_scaled,columns=['popularity', 'weighted_average', 'vote_average', 'revenue'])
movie_normalized_df.head()
```

```python
movies_cleaned[['normalized_weight_average','normalized_popularity','normalized_vote_average', 'noralized_revenue']]= movie_normalized_df
```

```python
movies_cleaned.head(20)
```

```python
movies_cleaned['score'] = movies_cleaned['normalized_weight_average']*0.25 + movies_cleaned['normalized_popularity']*0.25 + movies_cleaned['normalized_vote_average']*0.25 + movies_cleaned['noralized_revenue']*0.25
```

```python
movies_scored = movies_cleaned.sort_values(['score'], ascending=False)
movies_scored[['original_title', 'normalized_weight_average', 'normalized_popularity', 'normalized_vote_average','noralized_revenue','score']].head(20)
```
